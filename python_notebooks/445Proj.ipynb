```python
import altair as alt
import pandas as pd
```

```python

df = pd.read_csv("https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv")
df.head()
```

```python
df_subset = df[['state', 'season', 'date']].dropna(subset=['state'])

chart1 = alt.Chart(df_subset).mark_bar().encode(
    alt.X('state:N', sort='-y'),
    alt.Y('count()'),
    alt.Color('season:N')
).properties(
    title='Bigfoot Sightings by State',
    width=600,
    height=300
)

chart1
```

```python
# add a click selection on season
click = alt.selection_point(fields=['season'])

chart1 = alt.Chart(df_subset).mark_bar().encode(
    alt.X('state:N', sort='-y'),
    alt.Y('count()'),
    alt.Color('season:N')
).properties(
    title='Bigfoot Sightings by State',
    width=600,
    height=300
).transform_filter(
    click
)

chart2 = alt.Chart(df_subset).mark_bar().encode(
    alt.X('season:N'),
    alt.Y('count()'),
    alt.Color('season:N')
).properties(
    title='Click a Season to Filter',
    width=300,
    height=300
).add_params(
    click
)

chart = chart1 | chart2

chart
```
